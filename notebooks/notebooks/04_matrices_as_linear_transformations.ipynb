# 04.01 Why Graphics Needs Matrices
从 Chapter 04 开始，我们进入图形学中最核心的代数工具之一：

```text
matrix
```

矩阵不是一张数字表格而已。

在图形学里，矩阵首先应该被理解为：

```text
一种描述几何变换的方式。
```

更准确地说：

```text
矩阵描述了一个线性变换如何移动、缩放、旋转、拉伸或混合空间中的方向。
```

Chapter 04 的核心目标是理解：

```text
matrix as linear transformation
```

也就是：

```text
矩阵 = 对整个空间的一种规则化变形
```

## 04.01.01 Graphics Motivation

在图形学中，我们经常需要对点、向量、模型、相机和光线方向做变换。

例如：

```text
把模型从 local space 放到 world space；
把 world space 转到 camera / view space；
把 camera space 投影到 clip space；
把物体缩放、旋转、拉伸；
把法线方向从模型空间变换到世界空间；
把三角形顶点从一个坐标系转换到另一个坐标系。
```

这些操作如果分别写成很多独立公式，会非常混乱。

例如，一个 2D 点 p 先缩放，再旋转，最后再做某种坐标变换。

如果没有矩阵，我们需要分别写：

```text
x' = ...
y' = ...
```

但使用矩阵以后，我们可以统一写成：

$$
p'=Ap
$$

其中：

```text
p：原来的点或向量坐标
A：变换矩阵
p'：变换后的点或向量坐标
```

这就是矩阵在图形学中的第一层意义：

```text
矩阵把复杂的几何变换统一成一个代数对象。
```

## 04.01.02 Why Matrices Are Useful in Graphics

矩阵在图形学中有三个非常重要的作用。

第一，矩阵可以统一表示多种变换。

例如：

```text
scale
rotation
shear
reflection
basis change
local-to-world transformation
view transformation
projection transformation
```

第二，矩阵可以组合。

如果一个点先经过变换 B，再经过变换 A，那么结果可以写成：

$$
p'=A(Bp)
$$

由于矩阵乘法满足结合律，可以写成：

$$
p'=(AB)p
$$

也就是说，多个变换可以提前合成为一个矩阵：

$$
C=AB
$$

然后每个点只需要计算：

$$
p'=Cp
$$

这对图形学很重要，因为一个模型可能有成千上万个顶点。

第三，矩阵可以清楚地描述坐标系之间的关系。

在 column vector convention 下，矩阵的列向量有非常重要的几何意义：

```text
矩阵的每一列告诉我们一个 basis vector 被变换到了哪里。
```

这是本章最重要的直觉之一。

## 04.01.03 Matrix as a Transformation Rule

先考虑二维向量。

设：

$$
v=
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

设矩阵：

$$
A=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

矩阵作用在向量上：

$$
Av=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

结果是：

$$
Av=
\begin{bmatrix}
ax+by \\
cx+dy
\end{bmatrix}
$$

这说明矩阵 A 接收一个输入向量 v，然后输出一个新的向量 Av。

因此可以把矩阵 A 看成一个函数：

$$
T(v)=Av
$$

这里的 T 是一个 transformation。

也就是说：

```text
矩阵 A 定义了一个变换 T。
```

输入是：

```text
原来的向量 v
```

输出是：

```text
变换后的向量 T(v)
```

## 04.01.04 Columns Tell Where Basis Vectors Go

这一节最关键的结论是：

```text
在列向量约定下，矩阵的列向量就是标准基向量变换后的结果。
```

二维标准基向量是：

$$
e_1=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

$$
e_2=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

设：

$$
A=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

那么：

$$
Ae_1=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}=
\begin{bmatrix}
a \\
c
\end{bmatrix}
$$

这正好是矩阵 A 的第一列。

同理：

$$
Ae_2=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}=
\begin{bmatrix}
b \\
d
\end{bmatrix}
$$

这正好是矩阵 A 的第二列。

所以：

```text
A 的第一列 = e1 被变换后的位置
A 的第二列 = e2 被变换后的位置
```

更一般地：

```text
矩阵的第 i 列 = 第 i 个 basis vector 被变换后的结果
```

这是理解图形学矩阵的核心入口。

## 04.01.05 A Concrete Numerical Example

设有矩阵：

$$
A=
\begin{bmatrix}
2 & 1 \\
0 & 1
\end{bmatrix}
$$

它的第一列是：

$$
\begin{bmatrix}
2 \\
0
\end{bmatrix}
$$

它的第二列是：

$$
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

因此这个矩阵表示：

```text
原来的 e1 方向被变换到 (2, 0)
原来的 e2 方向被变换到 (1, 1)
```

也就是说：

$$
Ae_1=
\begin{bmatrix}
2 \\
0
\end{bmatrix}
$$

$$
Ae_2=
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

现在考虑向量：

$$
v=
\begin{bmatrix}
3 \\
2
\end{bmatrix}
$$

这个向量可以写成标准基的线性组合：

$$
v=3e_1+2e_2
$$

经过矩阵 A 变换后：

$$
Av=A(3e_1+2e_2)
$$

由于矩阵变换是线性的：

$$
Av=3Ae_1+2Ae_2
$$

代入：

$$
Av=3
\begin{bmatrix}
2 \\
0
\end{bmatrix}
+2
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

所以：

$$
Av=
\begin{bmatrix}
6 \\
0
\end{bmatrix}
+
\begin{bmatrix}
2 \\
2
\end{bmatrix}=
\begin{bmatrix}
8 \\
2
\end{bmatrix}
$$

这说明矩阵 A 不是单独移动某一个向量，而是规定了整个坐标网格如何变形。

向量 v 的新位置来自：

```text
变换后的 basis vectors 的线性组合。
```

## 04.01.06 The Geometric Meaning

原来的向量：

$$
v=
\begin{bmatrix}
3 \\
2
\end{bmatrix}
$$

可以理解为：

```text
沿 e1 方向走 3 份
沿 e2 方向走 2 份
```

矩阵 A 变换之后，新的规则是：

```text
原来的 e1 变成 Ae1
原来的 e2 变成 Ae2
```

所以变换后的向量是：

```text
沿 Ae1 方向走 3 份
沿 Ae2 方向走 2 份
```

用公式写就是：

$$
Av=3Ae_1+2Ae_2
$$

这就是为什么矩阵的列向量如此重要。

矩阵本质上告诉我们：

```text
旧的坐标轴方向被送到了哪里。
```

一旦知道 basis vectors 被送到了哪里，就知道整个空间中所有向量会被送到哪里。

## 04.01.08 Graphics Application

在图形学中，一个模型通常由很多顶点组成。

例如一个 2D 三角形有三个顶点：

$$
p_1=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

$$
p_2=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

$$
p_3=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

如果我们想对整个三角形做同一个线性变换 A，只需要对每个顶点都乘同一个矩阵：

$$
p_1'=Ap_1
$$

$$
p_2'=Ap_2
$$

$$
p_3'=Ap_3
$$

也就是说：

```text
同一个矩阵可以同时描述整个物体的变形规则。
```

这就是后续 model matrix 的基础。

不过要注意：

```text
普通 2D 或 3D 矩阵可以表示 scale、rotation、shear、reflection 等线性变换。
```

但是普通 2D 或 3D 线性矩阵不能直接表示 translation。

原因是 linear transformation 必须满足：

$$
T(0)=0
$$

而平移会把原点移动到别的位置。

例如：

$$
T(p)=p+t
$$

那么：

$$
T(0)=t
$$

如果 t 不是 zero vector，那么：

$$
T(0)\ne 0
$$

所以 translation 不是 linear transformation，而是 affine transformation。

后面我们会用 homogeneous coordinates 把 translation 也写进矩阵。

# 04.02 Basic Matrix Transformations in 2D
上一节我们已经建立了一个核心直觉：

```text
矩阵不是一张数字表，而是一种空间变换规则。
```

在 column vector convention 下：

```text
矩阵的列向量告诉我们标准基向量被变换到了哪里。
```

本节开始研究最常见的二维矩阵变换：

```text
scale
rotation
shear
reflection
```

这些变换是图形学中 model transformation 的基础。

本节核心目标是：

```text
理解每一种常见矩阵变换如何改变二维空间中的点、向量、坐标轴和图形。
```

## 04.02.01 Scaling Matrix

最简单的矩阵变换是缩放。

二维缩放矩阵通常写成：

$$
S=
\begin{bmatrix}
s_x & 0 \\
0 & s_y
\end{bmatrix}
$$

其中：

```text
s_x：x 方向的缩放比例
s_y：y 方向的缩放比例
```

给定二维向量：

$$
v=
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

缩放后的结果是：

$$
Sv=
\begin{bmatrix}
s_x & 0 \\
0 & s_y
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}=
\begin{bmatrix}
s_xx \\
s_yy
\end{bmatrix}
$$

所以 scaling matrix 的作用是：

```text
把 x 坐标乘以 s_x；
把 y 坐标乘以 s_y。
```

例如：

$$
S=
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
$$

表示：

```text
x 方向放大 2 倍；
y 方向放大 3 倍。
```

对向量：

$$
v=
\begin{bmatrix}
1 \\
2
\end{bmatrix}
$$

有：

$$
Sv=
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
\begin{bmatrix}
1 \\
2
\end{bmatrix}=
\begin{bmatrix}
2 \\
6
\end{bmatrix}
$$

## 04.02.02 Geometric Meaning of Scaling
从 basis vector 的角度看 scaling matrix。

二维标准基向量是：

$$
e_1=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

$$
e_2=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

缩放矩阵：

$$
S=
\begin{bmatrix}
s_x & 0 \\
0 & s_y
\end{bmatrix}
$$

作用在 e1 上：

$$
Se_1=
\begin{bmatrix}
s_x \\
0
\end{bmatrix}
$$

作用在 e2 上：

$$
Se_2=
\begin{bmatrix}
0 \\
s_y
\end{bmatrix}
$$

所以：

```text
e1 仍然沿 x 轴方向，只是长度变成 s_x；
e2 仍然沿 y 轴方向，只是长度变成 s_y。
```

这说明 scaling 不会改变坐标轴之间的夹角。

它只改变 basis vectors 的长度。

如果：

```text
s_x = s_y
```

那么这是 uniform scaling。

如果：

```text
s_x ≠ s_y
```

那么这是 non-uniform scaling。

图形学中经常使用 non-uniform scaling 来把模型压扁、拉长或改变比例。

## 04.02.03 Reflection Matrix

reflection 表示镜像变换。例如，关于 x 轴反射。

一个点：

$$
v=
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

关于 x 轴反射后变成：

$$
v'=
\begin{bmatrix}
x \\
-y
\end{bmatrix}
$$

对应的矩阵是：

$$
R_x=
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
$$

因为：

$$
R_xv=
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}=
\begin{bmatrix}
x \\
-y
\end{bmatrix}
$$

关于 y 轴反射，对应矩阵是：

$$
R_y=
\begin{bmatrix}
-1 & 0 \\
0 & 1
\end{bmatrix}
$$

因为它把：

```text
x 坐标变成相反数；
y 坐标保持不变。
```

所以：

$$
R_yv=
\begin{bmatrix}
-x \\
y
\end{bmatrix}
$$

关于原点反射，对应矩阵是：

$$
R_o=
\begin{bmatrix}
-1 & 0 \\
0 & -1
\end{bmatrix}
$$

它把：

$$
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

变成：

$$
\begin{bmatrix}
-x \\
-y
\end{bmatrix}
$$

## 04.02.04 Rotation Matrix

二维旋转矩阵是图形学中最重要的矩阵之一。

在标准右手坐标系中，如果使用 column vector convention，逆时针旋转角度 θ 的矩阵是：

$$
R(\theta)=
\begin{bmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{bmatrix}
$$

它的作用是：

```text
把二维向量绕原点逆时针旋转 θ。
```

为什么这个矩阵长这样？

关键仍然是看 basis vectors 被送到哪里。

标准基向量 e1 是：

$$
e_1=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

逆时针旋转 θ 后，e1 变成：

$$
R(\theta)e_1=
\begin{bmatrix}
\cos\theta \\
\sin\theta
\end{bmatrix}
$$

标准基向量 e2 是：

$$
e_2=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

逆时针旋转 θ 后，e2 变成：

$$
R(\theta)e_2=
\begin{bmatrix}
-\sin\theta \\
\cos\theta
\end{bmatrix}
$$

由于矩阵的列向量就是 basis vectors 变换后的结果，所以旋转矩阵的第一列是：

$$
\begin{bmatrix}
\cos\theta \\
\sin\theta
\end{bmatrix}
$$

第二列是：

$$
\begin{bmatrix}
-\sin\theta \\
\cos\theta
\end{bmatrix}
$$

因此：

$$
R(\theta)=
\begin{bmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{bmatrix}
$$

## 04.02.05 A 90-Degree Rotation Example

考虑逆时针旋转 90°。

此时：

$$
\cos 90^\circ=0
$$

$$
\sin 90^\circ=1
$$

所以旋转矩阵是：

$$
R(90^\circ)=
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
$$

对 e1：

$$
R(90^\circ)e_1=
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

这说明 x 轴方向被旋转到了 y 轴正方向。

对 e2：

$$
R(90^\circ)e_2=
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}=
\begin{bmatrix}
-1 \\
0
\end{bmatrix}
$$

这说明 y 轴方向被旋转到了 x 轴负方向。

对一般向量：

$$
v=
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

有：

$$
R(90^\circ)v=
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}=
\begin{bmatrix}
-y \\
x
\end{bmatrix}
$$

所以二维向量逆时针旋转 90° 的代数规则是：

```text
(x, y) -> (-y, x)
```

## 04.02.06 Shear Matrix

shear 通常翻译为剪切或错切。

它的作用是：

```text
让一个方向保持不变，同时让另一个方向根据它发生偏移。
```

常见的 x-direction shear matrix 是：

$$
H_x(k)=
\begin{bmatrix}
1 & k \\
0 & 1
\end{bmatrix}
$$

对向量：

$$
v=
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

有：

$$
H_x(k)v=
\begin{bmatrix}
1 & k \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}=
\begin{bmatrix}
x+ky \\
y
\end{bmatrix}
$$

也就是说：

```text
y 坐标保持不变；
x 坐标根据 y 的大小发生偏移。
```

如果 y 越大，那么 x 偏移越多。

所以 x-direction shear 会把图形“横向推斜”。

类似地，y-direction shear matrix 是：

$$
H_y(k)=
\begin{bmatrix}
1 & 0 \\
k & 1
\end{bmatrix}
$$

它的作用是：

$$
H_y(k)v=
\begin{bmatrix}
x \\
kx+y
\end{bmatrix}
$$

也就是说：

```text
x 坐标保持不变；
y 坐标根据 x 的大小发生偏移。
```

## 04.02.07 Geometric Meaning of Shear

继续用 basis vectors 理解 shear。

设：

$$
H_x(k)=
\begin{bmatrix}
1 & k \\
0 & 1
\end{bmatrix}
$$

第一列是：

$$
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

第二列是：

$$
\begin{bmatrix}
k \\
1
\end{bmatrix}
$$

所以：

$$
H_x(k)e_1=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

$$
H_x(k)e_2=
\begin{bmatrix}
k \\
1
\end{bmatrix}
$$

这说明：

```text
e1 不变；
e2 被推到 (k, 1)。
```

因此，x-direction shear 的几何意义是：

```text
保持 x 轴方向不变；
把 y 轴方向斜着推向右上方或左上方。
```

如果 k > 0，y 轴方向向右倾斜。

如果 k < 0，y 轴方向向左倾斜。

shear 不只是改变长度，它还会改变 basis vectors 之间的夹角。

因此，shear 会把正方形变成平行四边形。

## 04.02.08 Comparison of Four Basic Transformations

现在比较四种基本变换。

```text
scaling：
改变 basis vectors 的长度。

reflection：
翻转某些方向，可能改变 orientation。

rotation：
保持长度和角度，只改变方向。

shear：
保持某些方向不变，但改变夹角，把图形推斜。
```

更具体地说：

| Transformation | Preserve Length? | Preserve Angle? | Preserve Origin? | Common Graphics Use |
|---|---:|---:|---:|---|
| Scaling | No | If uniform, yes | Yes | Resize model |
| Reflection | Yes | Yes | Yes | Mirror object |
| Rotation | Yes | Yes | Yes | Rotate model/camera |
| Shear | No | No | Yes | Slant/deform geometry |

注意最后一列：

```text
这些二维矩阵都保持原点不变。
```

因为它们都是 linear transformations。

对于任何线性变换 T，都必须满足：

$$
T(0)=0
$$

所以普通 2D 矩阵可以表示 scaling、reflection、rotation、shear。

但是普通 2D 矩阵不能表示 translation。

translation 会在后续 homogeneous coordinates 中处理。

## 04.02.10 Transforming a Triangle

下面把一个二维三角形的三个顶点放在矩阵中。

设三角形顶点为：

$$
p_1=
\begin{bmatrix}

0
\end{bmatrix}
$$

$$

\begin{bmatrix}

0
\end{bmatrix}
$$

$$
p_3=
\begin{bmatrix}

1
\end{bmatrix}
$$

如果我们把每个顶点作为一列，可以写成：

$$

\begin{bmatrix}
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

其中：

```text
第一列是 p1；
第二列是 p2；
第三列是 p3。
```

$$
P'=AP
$$

这个写法非常接近图形学中的批量顶点变换。

$$

$$

$$

$$
P'=HP
$$

这说明：

```text
矩阵不只是作用在单个向量上，也可以批量作用在一组顶点上。
```

# 04.03 Matrix Composition and Transformation Order
上一节我们学习了四类基本二维矩阵变换：

```text
scaling
rotation
reflection
shear
```

本节进入图形学中非常关键的问题：

```text
多个矩阵变换如何组合？
```

以及：

```text
为什么矩阵变换的顺序很重要？
```

本节核心结论是：

```text
矩阵乘法通常不满足交换律。
```

也就是说：

$$
AB\ne BA
$$

在图形学中，这意味着：

```text
先旋转再缩放，通常不等于先缩放再旋转。
```

如果不理解这个问题，后面学习 model matrix、view matrix、projection matrix 时会非常容易混乱。

## 04.03.01 Composition of Transformations

假设有一个向量：

$$
v
$$

现在先对它做变换 B，再对结果做变换 A。

第一步：

$$
v'=Bv
$$

第二步：

$$
v''=Av'
$$

代入第一步：

$$
v''=A(Bv)
$$

由于矩阵乘法满足结合律：

$$
v''=(AB)v
$$

所以：

```text
先做 B，再做 A，可以合成一个矩阵 AB。
```

也就是说：

$$
C=AB
$$

然后：

$$
v''=Cv
$$

这就是 matrix composition。

它的意义是：

```text
多个几何变换可以提前合成为一个矩阵。
```

在图形学中，这非常重要。

因为一个模型可能有很多顶点，如果每个顶点都重复执行多个矩阵乘法，计算成本会比较高。

如果我们先把多个变换矩阵合成为一个矩阵，再作用到所有顶点上，就可以减少重复计算。

## 04.03.02 Right-to-Left Execution Order

本项目采用 column vector convention。也就是向量写成列向量：

$$
v=
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

矩阵作用在向量左边：

$$
v'=Av
$$

在这种约定下，如果有：

$$
v'=ABCv
$$

那么实际执行顺序是：

```text
先做 C；
再做 B；
最后做 A。
```

因为：

$$
v'=A(B(Cv))
$$

所以在 column vector convention 下：

```text
矩阵写在越右边，越先执行；
矩阵写在越左边，越后执行。
```

这是图形学矩阵顺序中最容易出错的地方。

例如：

$$
v'=RSv
$$

表示：

```text
先做 S scaling；
再做 R rotation。
```

而：

$$
v'=SRv
$$

表示：

```text
先做 R rotation；
再做 S scaling。
```

这两个结果通常不同。

## 04.03.04 Rotate Then Scale

现在换一个顺序。

仍然使用：

$$
v=
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

缩放矩阵：

$$
S=
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix}
$$

旋转矩阵：

$$
R=
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
$$

这次先旋转，再缩放。

第一步：

$$
Rv=
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
1 \\
1
\end{bmatrix}=
\begin{bmatrix}
-1 \\
1
\end{bmatrix}
$$

第二步：

$$
S(Rv)=
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
-1 \\
1
\end{bmatrix}=
\begin{bmatrix}
-2 \\
1
\end{bmatrix}
$$

所以：

$$
SRv=
\begin{bmatrix}
-2 \\
1
\end{bmatrix}
$$

这和上一节的结果不同：

$$
RSv=
\begin{bmatrix}
-1 \\
2
\end{bmatrix}
$$

而：

$$
SRv=
\begin{bmatrix}
-2 \\
1
\end{bmatrix}
$$

因此：

$$
RSv\ne SRv
$$

这说明：

```text
先缩放再旋转，不等于先旋转再缩放。
```

## 04.03.05 Matrix Multiplication Is Not Commutative

继续计算两个组合矩阵本身。

先计算：

$$
RS=
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix}
$$

结果是：

$$
RS=
\begin{bmatrix}
0 & -1 \\
2 & 0
\end{bmatrix}
$$

再计算：

$$
SR=
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
$$

结果是：

$$
SR=
\begin{bmatrix}
0 & -2 \\
1 & 0
\end{bmatrix}
$$

可以看到：

$$
RS\ne SR
$$

所以矩阵乘法通常不满足交换律。

但是矩阵乘法满足结合律：

$$
A(BC)=(AB)C
$$

这意味着：

```text
矩阵相乘时，不能随便交换顺序；
但是可以改变括号位置。
```

例如：

$$
A(B(Cv))=(ABC)v
$$

这是合法的。

但下面这个通常不合法：

$$
ABC=BAC
$$

因为交换了矩阵顺序。

## 04.03.06 Why Order Matters Geometrically

为什么顺序会影响结果？

核心原因是：

```text
每一次变换都会改变后续变换所作用的几何对象。
```

先缩放再旋转：

```text
先把物体沿原始 x 方向拉长；
再把拉长后的物体整体旋转。
```

先旋转再缩放：

```text
先把物体旋转；
再沿当前坐标系的 x 方向进行缩放。
```

如果缩放是 non-uniform scaling，那么不同方向的缩放比例不同。

所以：

```text
物体先朝哪个方向，被缩放的效果就会不同。
```

这就是为什么：

$$
RS\ne SR
$$

特别注意：

```text
如果是 uniform scaling，也就是所有方向缩放比例相同，那么它和 rotation 通常可以交换。
```

例如：

$$
S=
\begin{bmatrix}
s & 0 \\
0 & s
\end{bmatrix}
$$

这个矩阵等于：

$$
S=sI
$$

因此：

$$
RS=R(sI)=sR
$$

而：

$$
SR=(sI)R=sR
$$

所以：

$$
RS=SR
$$

但是 non-uniform scaling 一般不能和 rotation 交换。

## 04.03.07 Transforming an Entire Triangle

考虑一个三角形的三个顶点：

$$
p_1=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

$$
p_2=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

$$
p_3=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

把它们作为矩阵的列：

$$
P=
\begin{bmatrix}
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

如果先缩放再旋转，那么：

$$
P'=RSP
$$

执行顺序是：

```text
先 P；
然后 S；
最后 R。
```

如果先旋转再缩放，那么：

$$
P'=SRP
$$

执行顺序是：

```text
先 P；
然后 R；
最后 S。
```

这两个结果通常不同。

因此，在图形学中，写 transform pipeline 时必须明确：

```text
矩阵乘法的书写顺序；
矩阵实际执行顺序；
采用的是 column vector convention 还是 row vector convention。
```

本项目默认使用：

```text
column vector convention
```

因此：

```text
执行顺序从右往左。
```

## 04.03.08 Local Transformation vs World Transformation

在图形学中，还会遇到两个常见概念：

```text
local transformation
world transformation
```

假设一个模型点的 local coordinate 是：

$$
p_{\text{local}}
$$

模型矩阵是：

$$
M
$$

它把 local space 中的点变换到 world space：

$$
p_{\text{world}}=Mp_{\text{local}}
$$

现在有一个新的变换 A。

如果我们写：

$$
p_{\text{world}}'=AMp_{\text{local}}
$$

那么实际执行顺序是：

```text
先做 M；
再做 A。
```

这通常可以理解为：

```text
先把模型放到世界中；
再在世界坐标系下对它做 A。
```

因此：

$$
M_{\text{new}}=AM
$$

通常对应：

```text
在 world space 中追加一个变换。
```

如果我们写：

$$
p_{\text{world}}'=MAp_{\text{local}}
$$

那么实际执行顺序是：

```text
先做 A；
再做 M。
```

这通常可以理解为：

```text
先在模型自己的 local space 中对它做 A；
再把结果放到 world space。
```

因此：

$$
M_{\text{new}}=MA
$$

通常对应：

```text
在 local space 中追加一个变换。
```

这个区别在旋转物体时尤其重要。

例如：

```text
world rotation：绕世界坐标轴旋转；
local rotation：绕模型自身坐标轴旋转。
```

在 3D 图形学中，这个问题会更加重要。

## 04.03.11 Graphics Application

在图形学中，模型变换通常不是一个单独操作，而是多个操作组合而成。

例如，一个模型可能需要：

```text
先在 local space 中缩放；
再旋转；
最后移动到 world space 中某个位置。
```

如果暂时不考虑 translation，只看 scaling 和 rotation，那么可以写成：

$$
p'=RSp
$$

这表示：

```text
先做 scaling；
再做 rotation。
```

后面加入 homogeneous coordinates 后，我们会把 translation 也写成矩阵。

那时一个常见的 model matrix 可能长这样：

$$
M=TRS
$$

在 column vector convention 下：

$$
p_{\text{world}}=Mp_{\text{local}}
$$

也就是：

$$
p_{\text{world}}=TRSp_{\text{local}}
$$

实际执行顺序是：

```text
先做 S scaling；
再做 R rotation；
最后做 T translation。
```

所以：

```text
TRS 的执行顺序是 S -> R -> T。
```

这也是很多图形学教程里常见的 model matrix 组合顺序。

不过一定要注意：

```text
不同引擎、不同数学库、不同 shader convention 可能采用不同约定。
```

所以真正写代码时，必须确认：

```text
向量是 column vector 还是 row vector；
矩阵乘向量写成 Mv 还是 vM；
矩阵内存布局是 column-major 还是 row-major；
API 期望的矩阵顺序是什么。
```

本课程为了保持数学直觉清晰，统一采用：

```text
column vector convention
```
