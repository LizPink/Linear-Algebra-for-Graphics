# 02.01 Why Graphics Needs Dot Product

在 Chapter 01 中，我们已经学习了点、向量、坐标系、local space 和 world space。

从 Chapter 02 开始，我们要研究一个新的问题：

一个方向和另一个方向“有多一致”？

在图形学中，这个问题非常常见。例如：

1. 一个表面是否朝向光源？
2. 一个物体是否朝向相机？
3. 一个角色的移动方向和前方方向是否一致？
4. 一个点在某个方向上的投影长度是多少？
5. 两个方向是否垂直？
6. 一个三角形是否背对相机？

这些问题背后都需要 dot product。

## 02.01.01 Graphics Motivation

在计算机图形学中，很多对象本质上都是方向向量。

例如：

- surface normal：表面法线方向；
- light direction：光照方向；
- view direction：观察方向；
- camera forward：相机前方方向；
- object forward：物体朝向；
- movement direction：角色移动方向。

我们经常需要判断两个方向之间的关系：

- 它们是否大致同向？
- 它们是否垂直？
- 它们是否反向？
- 一个方向在另一个方向上有多少分量？

dot product 就是用来回答这些问题的基础工具。

## 02.01.02 A First Graphics Example: Surface Facing a Light

假设一个平面有一个法线方向 n，光从方向 l 照过来。

如果 n 和 l 越接近同向，表面越亮。

如果 n 和 l 垂直，光线擦过表面，亮度接近 0。

如果 n 和 l 反向，说明光在表面的背面，这个表面通常不被这束光照亮。

这个判断可以用 dot product 表示：

$$
\text{brightness}=n \cdot l
$$


如果结果为：

$$
n \cdot l>0
$$

说明两个方向夹角小于 90 度。

如果结果为：

$$
n \cdot l=0
$$

说明两个方向垂直。

如果结果为：

$$
n \cdot l<0
$$

说明两个方向夹角大于 90 度。

## 02.01.03 Dot Product as Alignment

dot product 可以先直观理解为：两个向量方向对齐程度的数值度量。

如果两个向量方向越一致，dot product 越大。如果两个向量互相垂直，dot product 等于 0。如果两个向量方向相反，dot product 为负。

所以在图形学中，dot product 经常被当成 alignment score。

## 02.01.04 Mathematical Definition

对于 3D 向量：

$$
u=
\begin{bmatrix}
u_x \\
u_y \\
u_z
\end{bmatrix}
$$

$$
v=
\begin{bmatrix}
v_x \\
v_y \\
v_z
\end{bmatrix}
$$

它们的 dot product 定义为：

$$
u \cdot v=u_xv_x+u_yv_y+u_zv_z
$$

dot product 的结果不是向量，而是一个标量。

## 02.01.05 Numerical Example

令：

$$
u=
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$

$$
v=
\begin{bmatrix}
0 \\
1 \\
0
\end{bmatrix}
$$

则：

$$
u \cdot v=1 \times 0+0 \times 1+0 \times 0=0
$$

所以 u 和 v 垂直。

这对应右手坐标系中的 x 轴方向和 y 轴方向。

## 02.01.06 Another Numerical Example

令：

$$
u=
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$

$$
v=
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$

则：

$$
u \cdot v=1 \times 1+0 \times 0+0 \times 0=1
$$

两个向量完全同向。

如果：

$$
w=
\begin{bmatrix}
-1 \\
0 \\
0
\end{bmatrix}
$$

则：

$$
u \cdot w=1 \times (-1)+0 \times 0+0 \times 0=-1
$$

说明 u 和 w 完全反向。

## 02.01.09 Dot Product and Unit Vectors

在图形学中，很多方向向量都需要 normalize。

如果 u 和 v 都是单位向量，那么：

$$
u \cdot v=\cos(\theta)
$$

其中 theta 是两个向量之间的夹角。

因此，对于单位向量：

- dot product 接近 1：方向几乎相同；
- dot product 接近 0：方向几乎垂直；
- dot product 接近 -1：方向几乎相反。

这就是为什么 lighting、camera facing、view direction 等计算中经常先 normalize direction。

## 02.01.10 Graphics Application: Lambert Lighting Preview

Lambert lighting 是最基础的漫反射光照模型。

它的核心思想是：表面越朝向光源，越亮。

如果 n 是单位法线方向，l 是单位光照方向（图形学中表示物体表面指向光源的方向），则漫反射强度可以写成：

$$
I=\max(0,n \cdot l)
$$

这里使用 max 是因为：

如果 n 和 l 的 dot product 为负，说明光在表面背后，不能贡献正的漫反射亮度。

所以实际亮度不会是负数，而是截断为 0。

```python
# 表面法线：朝向 +z
import numpy as np
n = np.array([0.0, 0.0, 1.0])

# 三个光照方向
light_front = np.array([0.0, 0.0, 1.0])
light_side = np.array([1.0, 0.0, 0.0])
light_back = np.array([0.0, 0.0, -1.0])

I_front = max(0.0, n @ light_front)
I_side = max(0.0, n @ light_side)
I_back = max(0.0, n @ light_back)

print("Front light intensity:", I_front)
print("Side light intensity:", I_side)
print("Back light intensity:", I_back)
```

## 02.01.11 Interpretation of the Lighting Example

在这个例子中：

$$
n=
\begin{bmatrix}
0 \\
0 \\
1
\end{bmatrix}
$$

如果光照方向（图形学中表示物体表面指向光源的方向）也是：

$$
l=
\begin{bmatrix}
0 \\
0 \\
1
\end{bmatrix}
$$

那么：

$$
n \cdot l=1
$$

表面完全朝向光源，亮度最大。

如果光照方向是：

$$
l=
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$

那么：

$$
n \cdot l=0
$$

光线从侧面擦过表面，亮度为 0。

如果光照方向是：

$$
l=
\begin{bmatrix}
0 \\
0 \\
-1
\end{bmatrix}
$$

那么：

$$
n \cdot l=-1
$$

光在表面背面，经过 max 截断后亮度为 0。
